### My Solution

```python
age = int(input("Enter your age: "))
day_type = input("Enter the day type (weekday/weekend): ")

if (age < 12):
  if (day_type == "weekday"):
    print("Ticket Price: 120 BDT\nEnjoy your movie!")
  elif (day_type == "weekend"):
    print("Ticket Price: 150 BDT\nEnjoy your movie!")
  else:
    print("Invalid input! Please enter either 'weekday' or 'weekend'.")

elif (12 <= age <= 17):
  if (day_type == "weekday"):
    print("Ticket Price: 200 BDT\nEnjoy your movie!")
  elif(day_type == "weekend"):
    print("Ticket Price: 250 BDT\nEnjoy your movie!")
  else:
    print("Invalid input! Please enter either 'weekday' or 'weekend'.")

elif (18 <= age <= 59):
  if (day_type == "weekday"):
    print("Ticket Price: 300 BDT\nEnjoy your movie!")
  elif(day_type == "weekend"):
    print("Ticket Price: 350 BDT\nEnjoy your movie!")
  else:
    print("Invalid input! Please enter either 'weekday' or 'weekend'.")

elif (age >= 60):
  if (day_type == "weekday"):
    print("Ticket Price: 180 BDT\nEnjoy your movie!")
  elif(day_type == "weekend"):
    print("Ticket Price: 200 BDT\nEnjoy your movie!")
  else:
    print("Invalid input! Please enter either 'weekday' or 'weekend'.")
```
