## ðŸ§  Problem Statement

Design a program that recommends items from a vending machine based on the inserted amount.

| Item | Price |
|------|--------|
| Water | 15 |
| Soda | 25 |
| Chips | 35 |
| Chocolate | 45 |
| Sandwich | 60 |

### Input Example
40


### Expected Output
You can buy:

*   Water
*   Soda
*   Chips


## ðŸ’¡ Thought Process

1. Use a dictionary to store items and prices.  
2. Loop through the items and print those â‰¤ amount.  
3. If no item matches, print "Insufficient balance."  
4. Optional: Find combinations that fit the budget exactly.


```python
def vending_machine(amount):
    items = {"Water": 15, "Soda": 25, "Chips": 35, "Chocolate": 45, "Sandwich": 60}
    affordable_items = [item for item, price in items.items() if price <= amount]

    if affordable_items:
        print("You can buy:")
        for item in affordable_items:
            print("-", item)
    else:
        print("Insufficient balance.")
```

```python
amount = int(input("Insert money: "))
vending_machine(amount)
```
